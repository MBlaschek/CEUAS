```python
import plotly
import pandas as pd 
import plotly.graph_objects as go
import plotly.express as px


CUON = '/users/staff/federico/GitHub/CEUAS_master_JULY2922/CEUAS/CEUAS/meta/inventory_comparison_2/code/station_configuration/CUON_station_configuration_extended.csv'
df = pd.read_csv(CUON, sep = '\t')
```

```python
def make_map(df):
    
    #df["color"] = "blue"
    
    
    fig = go.Figure()

    fig=px.scatter_geo(df,lon='longitude', lat='latitude',
                       opacity=0.8,
                       #projection="natural earth",
                       hover_data={'primary_id': True,
                                   'longitude':True,
                                   'latitude':True,
                                   'elevation':True,
                                   'station_name':True,
                                  },
                      range_color=[0,5000],
                      color = 'elevation',
                      title = 'Global CUON inventory',
                      color_continuous_scale=px.colors.diverging.BrBG_r )
                       

    fig.update_layout(
    #title="Map for the Database " + db,

    margin=dict(l=0, r=1, t=50, b=1),
    paper_bgcolor="white",
    width=1500,
    height=800)

    return fig 
```

```python
a = make_map(df)
```
