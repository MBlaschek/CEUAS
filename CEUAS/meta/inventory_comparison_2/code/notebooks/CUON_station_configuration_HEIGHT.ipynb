```python
def make_map(df):
    
    #df["color"] = "blue"
    
    
    fig = go.Figure()

    fig=px.scatter_geo(df,lon='longitude', lat='latitude',
                       opacity=0.8,
                       #projection="natural earth",
                       hover_data={'primary_id': True,
                                   'longitude':True,
                                   'latitude':True,
                                   'elevation':True,
                                   'station_name':True,
                                  },
                      range_color=[0,5000],
                      color = 'elevation',
                      title = 'Global CUON inventory - Station Elevation',
                      color_continuous_scale=px.colors.diverging.BrBG_r,
                      )
                       

    fig.update_layout(
    #title="Map for the Database " + db,

    margin=dict(l=0, r=1, t=50, b=1),
    paper_bgcolor="white",
    width=1500,
    height=800)

    fig.update_layout(
    #title="Plot Title",
    xaxis_title="",
    yaxis_title="",
    #legend_title="Legend Title",
    font=dict(
        #family="Courier New, monospace",
        size=20,
        #color="RebeccaPurple"
    ),
        
    coloraxis_colorbar=dict(
    title="Elevation [m]", )
    )

    
    return fig 
```

```python
a = make_map(df_r)
```
