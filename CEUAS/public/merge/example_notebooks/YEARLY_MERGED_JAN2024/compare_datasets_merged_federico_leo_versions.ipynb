```python
def read_data(f, date_min='19410101' , date_max='19410601'):
    """ Quickly reading some data """
    f = h5.File(f, 'r')
    ot = f['observations_table']

    dt = pd.to_datetime( f['recordtimestamp'][:] , unit='s',  origin=pd.Timestamp('1900-01-01') ).date 

    indices = np.where( ( dt < pd.Timestamp(date_max) ) & (dt > pd.Timestamp(date_min)) ) 

    #print(indices)
    imin, imax =  f['recordindex'][indices[0][0]] , f['recordindex'][indices[0][-1]]
                                                                     
    dic = {}
    for v in ['date_time' , 'z_coordinate' , 'z_coordinate_type' , 'observation_value' , 'observed_variable', 'source_id']:
        if v != 'source_id':
            dic[v] = ot[v][imin:imax   ]
        else:
            dic[v] = [ b''.join(a) for a in ot[v][imin:imax   ] ]
            

    df = pd.DataFrame.from_dict(dic)
    df['date_time'] = pd.to_datetime( df['date_time'][:] , unit='s',  origin=pd.Timestamp('1900-01-01') ) 
    
    df = df.dropna( subset=['observation_value'])
    return df 
```
