```python
unique_wmo = list(np.unique( dfu.wmo.values ))


compatible_sch = {}

for uw in unique_wmo:
    uw = int(uw)
    if uw not in mapping_wmo_to_sch.keys():
        continue 
        
    compatible_sch[uw] = []
    
    # unspecified WMO sensor 
    if uw in [90]: 
        continue
                
    dfr = dfu.loc[dfu.wmo.astype(int) == uw ]
    #print('\nLoop on WMO code ' , uw )
    

    print(dfr)
        
    for i,r in dfu.iterrows():
        find_sch , find_wmo = False, False
        
        wmo_com, sch_com = r.wmo_comment, r.sch_comment
        sch = r.sch
        
        if 'any' not in mapping_wmo_to_sch.keys():
            # check if ALL the flags for WMO and Sch appear in the comments
            for w in mapping_wmo_to_sch[uw]['wmo']:
                if w in wmo_com:
                    find_wmo = True
                else:
                    find_wmo = False
            
            for s in mapping_wmo_to_sch[uw]['sch']:
                if s in sch_com:
                    find_sch = True
                else:
                    find_sch = False    
        else:
            # here only one is sufficient 
            for w in mapping_wmo_to_sch[uw]['wmo']:
                if w in wmo_com:
                    find_wmo = True
         
            for s in mapping_wmo_to_sch[uw]['sch']:
                    if s in sch_com:
                        find_sch = True

        if find_wmo and find_sch:
            if sch not in compatible_sch[uw]:
                compatible_sch[uw].append(sch)
            
    
    
```
