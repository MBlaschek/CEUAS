```python
out = {}
out['station'] = []
out['header_table_ri'] = []
out['duplicates'] = []
out['header_table_duplicates'] = []


stats = glob.glob('/mnt/users/scratch/leo/scratch/converted_v9/newindex/0-*.nc')[:]
df_first = True
for conv_file in stats:
    sid = conv_file.split('_CEUAS_merged_v1.nc')[0][-5:]
    print(sid)
    out['station'].append(sid)
    with h5py.File(conv_file, 'r') as file:
        ht_ri = file['header_table']['report_id'][:]
        
    ht_ri = ht_ri.astype(object).sum(axis=1).astype(str)
    out['header_table_ri'].append(len(ht_ri))
    out['duplicates'].append(len(ht_ri) - len(np.unique(ht_ri)))
    
    u, c = np.unique(ht_ri, return_counts=True)
    dup = u[c > 1]
    out['header_table_duplicates'].append(str(dup))
df = pd.DataFrame.from_dict(out)
```
