```python
def datetime_to_seconds(dates, ref='1900-01-01T00:00:00'):
    """ from datetime64 to seconds since 1900-01-01 00:00:00"""
    return ((dates - np.datetime64(ref)) / np.timedelta64(1, 's')).astype(np.int64)

def seconds_to_datetime(seconds, ref='1900-01-01'):
    """ from seconds to datetime64 """
    seconds = np.asarray(seconds)
    return pd.to_datetime(seconds, unit='s', origin=ref)
```

```python
def compare_ingested_data(stat, year, mon):
    print(stat, year, mon)
    df = pd.read_csv('/users/staff/a1400070/CDS/cads-obs-catalogue-manager/tests/'+stat+'_test_ingested_data_'+year+mon+'.csv')
    df.date_time = seconds_to_datetime(df.date_time)

    df_cut = df[['z_coordinate', 'observed_variable', 'observation_value', 'date_time']]
    df_cut = df_cut[df_cut.observed_variable == 126]
    df_cut.date_time = pd.to_datetime(df_cut.date_time)
    print('ingestion data')
    display(df_cut)

    request={
        'date': year+mon+'01-'+year+mon+'31',
        'statid': stat,
        "format": "csv",
    #     "optional": ['latitude_displacement', 'longitude_displacement'],
    #     "pressure_level":['30000'],
        "observed_variable": [
            "air_temperature",
    #         "relative_humidity",
            #"u_component_of_wind",
            #"v_component_of_wind"
        ],
    }

    df = eua.vm_request_wrapper(request , overwrite=True, vm_url='http://127.0.0.1:8007')
    df.date_time = pd.to_datetime(df.date_time, utc=None).dt.tz_localize(None)
    print('cuon data')
    display(df) #= 


    missmatch = False
    for i in df.date_time.drop_duplicates():
        df_check = df[df.date_time == i]
        for j in df_check.z_coordinate:
            df_test = df_cut[df_cut.date_time == i]
    #         display(df_test)
            df_test = df_test[df_test.z_coordinate == j]
            if len(df_test) < 1:
                print(i,j)
                missmatch = True
        if missmatch:
            display(df_check)
            display(df_test)
            missmatch = False
    print(' \n \n')
```
